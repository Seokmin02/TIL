2. 덧셈횟수 n번

```python
A  = [1,2,3,4,5]
output = 0
for a in A:
    output += a
print(output)
```

3. 덧셈횟수 n^2번

```python
A = [ 
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
output = 0
for lst in A:
    for i in lst:
        output += i
print(output)
```

n = 100,000,000 일때  

8n^2 + 4 = 80000000000000004  

3n + 1241= 300001241  

1 = 1  

### 숫자가 증가하는 비율에 영향을 주는 것은 최고차 항 뿐이다.

#### Big-O 표기법
- 최고차 항만 남기고
- 최고차 항의 계수를 제거하면

8n^2 + 4 => O(n^2)

3n + 1241 => O(n)

1 = 1  => O(1)


n! + n^2 + nlogn + 1
=> O(n!)

#### 프론티어 슈퍼컴퓨터
연산속도는 1.12엑사 플롭 급임 => 1초에 112경 번의 부동소수점 연산이 가능.  
이걸로 O(n!) 연산을 해본다면?  

1경 = 10^16  
n = 32일때 O(n!) 연산을 한다면  
10 ^35 / 10^18 = 10^17초

```python
print(f"{10 ** 17 / (60 * 60 *24 *365)}년")
```

위처럼 O(n!) 복잡도는 현실적으로 활용불가능한 속도임
- 데이터 수가 적을때는 최대 n^2
- 데이터 수가 많을때는 최대 nlogn 정도까지 나와야 현실적으로 사용 가능한 알고리즘
